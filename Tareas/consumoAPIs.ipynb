#Tarjeta Cr√©dito

```python
import requests
import json
import pandas as pd

url = "https://fakerapi.it/api/v1/credit_cards?_quantity=1"
respuesta = requests.get(url)
datos = respuesta.json()
texto = json.loads(respuesta.text)
tabla = texto['data']
pd.DataFrame(tabla)
```
